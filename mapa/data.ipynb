```python
# Lista de columnas a conservar
columnas_a_conservar = ['nom_estab', 'nombre_act', 'municipio', 'localidad', 'latitud', 'longitud']

# Eliminar todas las demás columnas
df_nuevo = df[columnas_a_conservar]
df_nuevo.info()
df_nuevo.head()

# Lista de municipios de interés
#municipios_interes = ["Zumpango", "Tequixquiac", "Apaxco"]

# Filtrar el DataFrame para conservar solo esos municipios
#df_nuevo = df_nuevo[df_nuevo['municipio'].isin(municipios_interes)]

# Verificar que quedó correcto
#df_nuevo.info()
#df_nuevo['municipio'].unique()
```

```python
#export df_nuevo to csv
df_nuevo.to_csv('denue_inegi_15_2_reducido.csv', index=False)
```

```python
import geopandas as gpd
from shapely.geometry import Point

#Crear geometria
# Leaflet y GeoJSON usan (longitud, latitud)
geometry = [Point(xy) for xy in zip(df_nuevo['longitud'], df_nuevo['latitud'])]

#  Crear GeoDataFrame
gdf = gpd.GeoDataFrame(df_nuevo, geometry=geometry)

# Definir el sistema de coordenadas (CRS)
gdf.set_crs(epsg=4326, inplace=True)  # EPSG:4326 = lat/lon estándar

# Seleccionar solo columnas necesarias para GeoJSON
gdf = gdf[['nom_estab', 'nombre_act', 'municipio', 'localidad', 'geometry']]

# Exportar a GeoJSON
gdf.to_file("negocios_edomex.geojson", driver='GeoJSON')
```
